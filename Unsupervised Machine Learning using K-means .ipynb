```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits.data.shape

```

```python
digits.target_names
```

```python
digits.data[0].reshape(8,8)
```

```python
plt.imshow(digits.data[0].reshape(8,8),cmap=plt.cm.binary)
```

```python
digits.target[0]
```

```python
cost= []
models = []

for i in range(1,20):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(digits.data)
    models.append(kmeans)
    print (kmeans.score(digits.data))
    cost.append(-1 * kmeans.score(digits.data))
```

```python
cost
```

```python
plt.plot(range(1,20),cost)
```

```python
best_model = models[9]
```

```python
clusters = best_model.predict(digits.data)
```

```python
clusters.shape
```

```python
clusters = np.where(clusters == 0)
```

```python
digits.target[clusters]
```

```python
mode(digits.target[clusters])
```
